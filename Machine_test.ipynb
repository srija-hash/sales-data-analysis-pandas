```python
#Task 1: Data Understanding
import os
import pandas as pd 
df = pd.read_csv("C:\\Users\\rjith\\Machine_test.csv")
df = df.iloc[:,0].str.split(',',expand = True)
df.columns = ['Customer_ID','Name','Age','Gender','City','Product','Price','Quantity','Payment_Mode','Purchase_Date'] 
print(df)
print(df.shape)
print('Column names:',df.columns)
print('Data types:',df.dtypes)
print(df.isnull().sum())
#Task 2: Data Cleaning
df = df.replace(['', ' '], pd.NA) # Fix empty values → NA
df['Age'] = df['Age'].replace(['', ' '], pd.NA)
df['Age'] = df['Age'].fillna(0)
df['Age']=df['Age'].astype('float')
df = df.drop_duplicates() # Remove duplicates
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date']) # Convert date
df= df.rename(columns={'Payment_Mode':'PaymentType'}) # Rename column
print('Task 2:\n',df)
#Task 3: Feature Engineering
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Total_Amount'] = df['Price'] * df['Quantity'] #Total amount
df['Age_group'] = pd.cut(df['Age'], bins=[0,18,30,55,100],labels=['Teen','Young','Adult','Senior']) # Age Group
df['Gender'] = df['Gender'].astype('category') # Optimize data types
df['City'] = df['City'].astype('category')
df['Product'] = df['Product'].astype('category')
print('Task 3:',df)
# Task 4: Data Filtering
# Convert columns to numeric FIRST
df['Age'] = pd.to_numeric(df['Age'])
df['Price'] = pd.to_numeric(df['Price'])
df['Quantity'] = pd.to_numeric(df['Quantity'])
df_kochi = df[df['City'] == 'Kochi']
df_filter = df[(df['Age'] > 25) & (df['Product'] == 'Laptop')]
df_query = df.query("City == 'Chennai' and Price > 20000")
print('Task 4:',df)
# Task 5: Data Analysis
print('Task 5:')
df['Price'] = df['Price'].astype('float')
avg_price = df['Price'].mean()
total_quan = df['Quantity'].sum()
print('Average:',avg_price)
print('Total quantity',total_quan)
# Task 6: Advanced Analysis
print('Task 6:')
pivot = pd.pivot_table(df, values='Total_Amount', index='City', columns = 'Product', aggfunc='sum')
print('Pivot table:\n',pivot)
ct=pd.crosstab(df['Gender'], df['Product'])
print('Crosstab:\n',ct)
df['Year'] = df['Purchase_Date'].dt.year
df['Month'] = df['Purchase_Date'].dt.month
#Task 7: Data Integration
print('Task 7:\n')
region = pd.DataFrame({ 'City':['Chennai','Kochi','Hyderabad','Trivandrum'], 'Region': ['South','North','East','West']})
print(region)
merged = pd.merge(df,region, on='City', how='inner')
print('Merge:\n',merged)
#Task 8: Data Export
print('Task 8:\n')
df.to_csv('final_output.csv')
df.to_excel('final_output.xlsx', sheet_name = 'Sales', index=False)
df.to_json('final_output.json', date_format='iso')
file_path = os.path.abspath('final_output.json')
print(f'File Saved at:{file_path}')
#Task 9: Aggregation and Transformation
print('Task 9:')
df['Price_Category'] = df['Price'].apply(
    lambda x: 'High' if x > 20000 else 'Low'
)
print(df)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Total_Amount'] = df['Price'] * df['Quantity']
agg = df.groupby('City').agg({'Total_Amount': ['sum','mean','max'],'Quantity': ['sum','mean']})
print(df['Customer_ID'].is_unique)
```
